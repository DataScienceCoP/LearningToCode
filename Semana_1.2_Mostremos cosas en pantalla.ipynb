# Mostrando los tipos de datos

Como ya mencionamos Python reconoce distintos tipos de datos. Por ejemplo, n칰meros enteros, n칰meros con decimales, booleano (es cuando el resultado es cero o uno), textos, fechas, listas, diccionarios, y archivos.

Por ejemplo:
*   Enteros o *integers*: -1, 0, 1
*   Flotantes o *floats*: -1.3, 0.005, 1.333333
*   Booleanos o *booleans*: 1, 0, True, False
*   Listas o *lists*: [1, "dos", 3, 4.1]
*   Textos o *strings*: "2", 'hola', "칄l dijo que 'comprar칤a' la cena."

Los *strings* pueden ser delimitados por:<br>
*   ' ' ap칩strofe sencilla

*   ''' '''triple ap칩strofe<br>
*   """ """ triple comilla

Adem치s, los *strings* son inmutables; es decir, que no cambian por si solos, pero los operadores (in, +, y **)* y funciones (*len*, *min*, y *max*) pueden dar un nuevo *string* a partir del original.

````python
# Ejemplo un texto con tab \t (este s칤mbolo significa tabulaci칩n o espaciado y es utilizado en diversos lenguajes para 
# indicarle a la computadora cuando un texto lleva un espaciado similar al apretar la tecla Tab)
# Definamos una variable x con alg칰n texto
x = "Soy fan de las pel칤culas de \t  \tDC"
print(x) # Utilizamos la funci칩n print para imprimir o mostrar un resultado. En este caso, mostramos el contenido de la variable x

# 춰Muy bien! Ahora usemos la funcion split para separar el contenido de la variable x 
y = x.split()
print(y)
````

Otros tipos de datos pueden ser convertidos en texto e incluso le podemos dar el formato que deseemos. Veamos el siguiente ejemplo.

````python
# Definamos las siguientes variables 
e = 2.718125
z = [1, "dos", 3, 4.0, ["A", "b"]]

print("La constante e es: ", e, " y la lista z es: ", z)
````

````python
# Sin embargo, en Python, generalmente hay varias maneras de llegar al mismo resultado
print("La constante e es: {:.3f} y la lista z es: {}".format(e, z[-2]))
````

````python
# Notemos que en este ejemplo no cambiamos la variable, sino que seleccionamos un valor de la lista y modificamos el formato
print("La constante e es: {:.2f} y la lista z es: {:.4f}".format(e, z[1]))
````
