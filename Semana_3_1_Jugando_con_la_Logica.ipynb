# OPERADORES DE COMPARACION

Los operadores de comparación nos ayudan a comparar un valor con otro. Cuando usamos operadores de comparación, nos devuelven una respuesta verdadera o falsa (TRUE / FALSE) conocida como tipo booleano. Los operadores de comparación y los booleanos son muy importantes porque nos ayudan a tomar decisiones en nuestro código.

# MAYOR QUE (>)

Cuando lo usa, la computadora decide si el valor en el lado izquierdo del símbolo> es mayor que el valor en el lado derecho del símbolo.

```python
3>4
```

```python
10>5
```

# OPERADORES LOGICOS
se utilizan para ayudarnos a comparar operandos verdaderos o falsos. Estos son muy útiles porque pueden hacer que nuestra decisión de hacer reglas sea más compleja. Lo que significa un código más inteligente. Hay tres operadores lógicos principales
#AND
El operador AND comprueba que los valores a la derecha e izquierda sean ambos TRUE. Si hay un punto en nuestro código que solo debe ejecutarse cuando se cumplen dos condiciones, debemos usar el operador AND.
#OR
El operador OR comprueba para asegurarse de que al menos un valor que se está comparando sea TRUE.
#NOT
El operador NOT comprueba para asegurarse de que el valor que se está comparando sea FALSE.


```python
A = True
B = False

print('A and B is', A and B)

print('A or B is', A or B)

print('not A is', not B)
```

```python
if 590 == 110 or 99 == 99 :
  print("at least one of the statements is true")
```

```python
a = 10
  
if not a:
    print("Boolean value of a is True")
else:
    print("10 is divisible by either 3 or 5")
```

# IF STATEMENT

Es un bloque de codigo que nos permite controlar el camino que la computadora cuando ejecuta el codigo

```python
clima = input("""
Como esta el clima?
- soleado
- lluvioso
- nublado
""")

if clima == "soleado":
    print ("El dia esta soleado, podemos ir al parque☀️🏞️🚶 ")
```

```python
num = 10
if num > 0:
    print(num, "is a positive number.")
print("This is always printed.")

num = -5
if num > 0:
    print(num, "is a positive number.")
print("This is also always printed.")
```

# ELSE

Es una forma de ejecutar código si la condicion if no se cumple

```python
clima = input("""
Como esta el clima?
- soleado
- lluvioso
- nublado
""")

if clima == "soleado":
    print ("El dia esta soleado, podemos ir al parque☀️🏞️🚶 ")
else:
    print("No conozco ese clima 🤔")
```

```python
edad = 18

if edad >= 18:
  print("Eres mayor de edad!")
else:
  print("Todavia eres menor de edad!")
```

# ELIF

Con esta condición `elif` agregamos la posibilidad de incluir una condicion si la primera no se cumple.

```python
clima = input("""
Como esta el clima?
- soleado
- lluvioso
- nublado
""")

if clima == "soleado":
    print ("El dia esta soleado, podemos ir al parque☀️🏞️🚶 ")
elif clima == "lluvioso":
    print ("El dia esta lluvioso, mejor nos quedamos en la casa 🏠⛈️")
elif clima == "nublado":
    print("El dia esta nublado, ten cuidado con los rayos ⚡")
else:
    print("No conozco ese clima 🤔")
```

```python
edad = 18

if edad >= 18:
  print("Eres mayor de edad!")
elif edad >= 0:
  print("Todavia eres menor de edad!")
else:
  print("Edad inválida!")
```

1. Escribe un programa en Python para convertir temperaturas a y desde celsius, fahrenheit.

[Fórmula : c/5 = f-32/9 [ donde c = temperatura en celsius y f = temperatura en fahrenheit ]

Resultado esperado :
60°C es 140 en Fahrenheit
45°F es 7 en Celsius


```python
temp = input("Input the  temperature you like to convert? (e.g., 45F, 102C etc.) : ")
degree = int(temp[:-1])
i_convention = temp[-1]

if i_convention.upper() == "C":
  result = int(round((9 * degree) / 5 + 32))
  o_convention = "Fahrenheit"
elif i_convention.upper() == "F":
  result = int(round((degree - 32) * 5 / 9))
  o_convention = "Celsius"
else:
  print("Input proper convention.")
  quit()
print("The temperature in", o_convention, "is", result, "degrees.")
```


2. Escribe un programa en Python para calcular la edad de un perro en años de perro
Nota: Durante los dos primeros años, un año canino equivale a 10,5 años humanos. Después, cada año canino equivale a 4 años humanos.
Resultado esperado:

Introduzca la edad de un perro en años humanos: 15                                    
La edad del perro en años de perro es 73

```python
h_age = int(input("Input a dog's age in human years: "))

if h_age < 0:
	print("Age must be positive number.")
	exit()
elif h_age <= 2:
	d_age = h_age * 10.5
else:
	d_age = 21 + (h_age - 2)*4

print("The dog's age in dog's years is", d_age)
```
