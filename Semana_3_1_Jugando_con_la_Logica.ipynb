# OPERADORES DE COMPARACION

Los operadores de comparaci√≥n nos ayudan a comparar un valor con otro. Cuando usamos operadores de comparaci√≥n, nos devuelven una respuesta verdadera o falsa (TRUE / FALSE) conocida como tipo booleano. Los operadores de comparaci√≥n y los booleanos son muy importantes porque nos ayudan a tomar decisiones en nuestro c√≥digo.

Cuando lo usa, la computadora decide si el valor en el lado izquierdo del s√≠mbolo> es mayor que el valor en el lado derecho del s√≠mbolo.

```python
3>4
```

```python
10>5
```

# OPERADORES LOGICOS
se utilizan para ayudarnos a comparar operandos verdaderos o falsos. Estos son muy √∫tiles porque pueden hacer que nuestra decisi√≥n de hacer reglas sea m√°s compleja. Lo que significa un c√≥digo m√°s inteligente. Hay tres operadores l√≥gicos principales
#AND
El operador AND comprueba que los valores a la derecha e izquierda sean ambos TRUE. Si hay un punto en nuestro c√≥digo que solo debe ejecutarse cuando se cumplen dos condiciones, debemos usar el operador AND.
#OR
El operador OR comprueba para asegurarse de que al menos un valor que se est√° comparando sea TRUE.
#NOT
El operador NOT comprueba para asegurarse de que el valor que se est√° comparando sea FALSE.


```python
A = True
B = False

print('A and B is', A and B)

print('A or B is', A or B)

print('not A is', not B)
```

```python
if 590 == 110 or 99 == 99 :
    print("at least one of the statements is true")
```

```python
a = 10
  
if not a:
    print("Boolean value of a is True")
else:
    print("10 is divisible by either 3 or 5")
```

# ELSE

Es una forma de ejecutar c√≥digo si la condicion if no se cumple

```python
clima = input("""
Como esta el clima?
- soleado
- lluvioso
- nublado
""")

if clima == "soleado":
    print ("El dia esta soleado, podemos ir al parqueüåûüèûÔ∏èüö∂ ")
else:
    print("No conozco ese clima ü§î")
```

```python
edad = 18

if edad >= 18:
    print("Eres mayor de edad!")
else:
    print("Todavia eres menor de edad!")
```

# ELIF

Con esta condici√≥n `elif` agregamos la posibilidad de incluir una condicion si la primera no se cumple.

```python
clima = input("""
Como esta el clima?
- soleado
- lluvioso
- nublado
""")

if clima == "soleado":
    print ("El dia esta soleado, podemos ir al parqueüåûüèûÔ∏èüö∂ ")
elif clima == "lluvioso":
    print ("El dia esta lluvioso, mejor nos quedamos en la casa üè†‚õàÔ∏è")
elif clima == "nublado":
    print("El dia esta nublado, ten cuidado con los rayos ‚ö°")
else:
    print("No conozco ese clima ü§î")
```

```python
edad = 18

if edad >= 18:
    print("Eres mayor de edad!")
elif edad >= 0:
    print("Todavia eres menor de edad!")
else:
    print("Edad inv√°lida!")
```
